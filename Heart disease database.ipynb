# SPLIT DATA

```python
from sklearn.model_selection import train_test_split
x = df.drop('target',axis=1) #in this we take only independent data and drop dependent data  and x is x-axis which show all data except dependent data
y = df['target']   #in this we take only dependent data and y is y-axis which show dependent data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
print('x_train size:{}, x_test size:{}'.format(x_train.shape,x_test.shape))
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler  #it brings the the data into one range
sc = StandardScaler()
# fit scaler on training data
x_train = sc.fit_transform(x_train)
#tranform testing data
x_test = sc.fit_transform(x_test)
#x_scaled= (x - mesan(x)/std(x))
#and we take data only of independet data
```

```python
MSE_Score = []
R2_Score = []
Algorithm = []
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

# LOGISTIC REGRESSION

```python
Algorithm.append('Logistic Regression')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg_model = LogisticRegression()
reg_model.fit(x_train,y_train)
y_predict = reg_model.predict(x_train)
```

```python
r2_score(y_train,y_predict)
```

```python
y_testpredict = reg_model.predict(x_test)
```

```python
r2_score(y_test,y_testpredict)
```

```python
MSE_Score.append(mean_squared_error(y_test, y_testpredict))
R2_Score.append(r2_score(y_test, y_testpredict))
```

```python
for i in Algorithm, MSE_Score, R2_Score:
    print(i,end=',')
```

# DECISION TREE

```python
Algorithm.append('Decision Tree')
```

```python
r2_score(y_train,y1_predict)
```

```python
y1_testpredict=dt.predict(x_test)
```

```python
for i in Algorithm, MSE_Score, R2_Score:
    print(i,end=',')
```
